```python
### API Serveice Name 
# Ref. https://developers.google.com/youtube/v3/code_samples/python?hl=ja#search_by_keyword
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"

### Read api key from secret.json (YouTube Data)
# Note: service credentials from GCP is not required in case of Youtube API
import json
with open("secret-gcp-api-ytd.json") as f:
    secret = json.load(f)
DEVELOPER_KEY = secret["KEY"]

### Authenticate & Instantiate the client
youtube=build(
            YOUTUBE_API_SERVICE_NAME, 
            YOUTUBE_API_VERSION,
            developerKey=DEVELOPER_KEY
        )
```

```python
### Define video_search function to get response from Youtube API 
# Note: "youtube" is an instantiate client made above 
def video_search(youtube, query, max_result):

    # Call the search.list method to retrieve results matching the specified params
    response = youtube.search().list(
                    q = query,
                    part = "id,snippet",
                    maxResults = max_result,
                    order = "viewCount",
                    type = "video"
                ).execute()
    
    # Get results
    response = video_search(youtube, query, max_result)
    items = response["items"]

    # Extract the necesary items and render in beauty by pandas
    items_id = []
    for item in items:
        item_id = {}
        item_id["video_Id"] = item["id"]["videoId"]
        item_id["channel_Id"] = item["snippet"]["channelId"]
        items_id.append(item_id)

    import pandas as pd
    df_video =pd.DataFrame(items_id)

    return df_video
```

```python
### Prapre params for video_search
query = "Python 自動化"
max_result = 50

### Return result
result = video_search(youtube, query, max_result)
```
